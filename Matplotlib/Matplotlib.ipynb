```python
import pandas as pd 
d = pd.read_csv("AAPL.csv")
d.head()
```

```python
# We will load the pyplot submodule of Matplotlib so that we can draw our plots. 
# The pyplot module contains all of the relevant methods we will need to create plots and style them. 
# We will use the conventional alias plt. 
# We will also load in pandas, numpy, and datetime for future parts of this tutorial.
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
plt.plot(d['Date'], d['Close'])
```

```python
plt.plot(d['Date'], d['Open'])
plt.plot(d['Date'], d['Close'])


plt.show()
```

```python
plt.plot(d['Date'], d['Open'], label='Open')
plt.plot(d['Date'], d['Close'], label='Close')
plt.legend()
plt.show()
```

```python
d['Date'] = pd.to_datetime(d.Date, format='%Y-%m-%d')
```

```python
plt.bar(d['Date'], height = d['Close'])
plt.show()
```

```python
d['Date']
```

```python
plt.scatter(d['Open'], d['Close'])
plt.show()
```

```python
z = np.polyfit(d['Open'], d['Close'], 1)
p = np.poly1d(z)


plt.scatter(d['Open'], d['Close'])
plt.plot(d['Open'], p(d['Open']))
plt.show()
```

```python
plt.plot(d['Date'], d['Open'], color='black')
plt.plot(d['Date'], d['Close'], color='red')
plt.show()
```

```python
plt.scatter(d["Open"], d["Close"], c=d['Close'], cmap=plt.cm.plasma)
plt.show()
```

```python
plt.scatter(d["Open"], d["Close"], c=d['Close'], cmap=plt.cm.plasma)
plt.savefig('AAPL Stock Price - Open vs Close.png')
```
