```python
import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
trends = TrendReq()
```

```python
trends.build_payload(kw_list=["Virat Kohli"])
data = trends.interest_by_region()
data = data.sort_values(by="Virat Kohli", ascending=False)
data = data.head(10)
print(data)
```

```python
data.reset_index().plot(x="geoName", y="Virat Kohli", figsize=(15, 12), kind="bar")
plt.style.use('fivethirtyeight')
```

```python
data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=["Virat Kohli"])
data = data.interest_over_time()
fig, ax = plt.subplots(figsize=(15, 12))
data['Virat Kohli'].plot()
plt.style.use('fivethirtyeight')
plt.title('Total Google Searches for Machine Learning', 
          fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()
```
